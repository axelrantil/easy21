```python
from monte_carlo import monte_carlo_control
from sarsa import sarsa_lfa_control
from utils import plot_V
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
from tqdm import tqdm
```

```python
Q_opt, V = monte_carlo_control(N0=100, episodes=100000)
```

```python
mse_dict = {}
v_dict = {}
l = np.linspace(0,1,11)
mse_list = []
for lambd in tqdm(l):
    Q, V, mse = sarsa_lfa_control(episodes=10000, lambd=lambd, Q_opt=Q_opt)
    mse_dict[lambd] = mse
    v_dict[lambd] = V
    mse_list.append(mse[-1])
```

```python
for k in mse_dict.keys():
    #if k*10%2==0:
    plt.plot(mse_dict[k], label="Lambda = " + str(k))
plt.legend(loc='upper right')
plt.xlabel('Episodes')
plt.ylabel('MSE')
plt.show()
```
